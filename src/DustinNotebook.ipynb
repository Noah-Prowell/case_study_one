```python
# grab just the US in a dataframe
df_US = df[df['Country Name'] == 'United States'].transpose()
df_US.columns = df_US.iloc[3]
```

```python
# grab just the UK in a dataframe
df_UK = df[df['Country Name'] == 'United Kingdom'].transpose()
df_UK.columns = df_UK.iloc[3]
```

```python
#getting just the two columns we want (for both data frames)
cols = ['SP.POP.TOTL', 'SP.DYN.CDRT.IN', 'SP.DYN.CBRT.IN']

new_df_UK = pd.DataFrame(df_UK[cols])
new_df_US = pd.DataFrame(df_US[cols])

new_df_US.columns = ['Population', 'Death', 'Birth Rate']
new_df_UK.columns = ['Population', 'Death', 'Birth Rate']

new_df_US
```

```python
merged_df = new_df_US.join(new_df_UK, lsuffix='_US', rsuffix='_UK')
merged_df.drop(merged_df.index[0], inplace=True)
merged_df.drop(merged_df.index[0], inplace=True)
merged_df.drop(merged_df.index[0], inplace=True)
merged_df.drop(merged_df.index[0], inplace=True)
merged_df.drop(merged_df.index[-1], inplace=True)
merged_df.drop(merged_df.index[-1], inplace=True)

merged_df
```

```python
#slicing individual data segments for plotting

df_US_pop = merged_df.iloc[0:, 0:1]
df_UK_pop = merged_df.iloc[0:, 0:1]

df_US_death = merged_df.iloc[0:, 1:2]
df_UK_death = merged_df.iloc[0:, 1:2]

df_US_birth = merged_df.iloc[0:, 2:3]
df_UK_lit = merged_df.iloc[0:, 2:3]

df_UK_death
```

```python
years = [x for x in range(1960, 2015)]

fig1, axs1 = plt.subplots(2,1, figsize=(10,4))

axs1[0].plot(years, df_US_pop)
axs1[0].set_title('US Population (hundreds of millions)')
#axs1[0].set_ybound(100000000,350000000)
#axs1[0].set_yticks([1,2,3,4],[100000000,200000000,300000000,400000000])

axs1[1].plot(years, df_UK_pop)
axs1[1].set_title('UK Population (tens of millions)')

plt.tight_layout()
plt.savefig('population.png')
```

```python
fig2, axs2 = plt.subplots(2,1, figsize=(10,4))

axs2[0].plot(years, df_US_death, c='r')
axs2[0].set_title('US Death Rate (per 1,000 people)')

axs2[1].plot(years, df_UK_death, c='r')
axs2[1].set_title('UK Death Rate (per 1,000 people)')

plt.tight_layout()
plt.savefig('DeathRate.png')
```

```python
fig3, axs3 = plt.subplots(2,1, figsize=(10,4))

axs3[0].plot(years, df_US_lit, c='g')
axs3[0].set_title('US Birth Rate (per 1,000 people)')

axs3[1].plot(years, df_UK_lit, c='g')
axs3[1].set_title('UK Birth Rate (per 1,000 people)')

plt.tight_layout()
plt.savefig('BirthRate.png')
```

```python
axs3[0].scatter(years, df_US_death, c='r')
axs3[0].set_title('US Death')

axs3[1].scatter(years, df_UK_death, c='r')
axs3[1].set_title('UK Death')

plt.tight_layout()
```
